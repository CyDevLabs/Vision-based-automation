```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
im = cv.imread('C:/Users/Madara/Pictures/1.jpg', cv.IMREAD_COLOR)
hist_b = cv.calcHist([im], [0], None, [256], [0,256])
hist_g = cv.calcHist([im], [1], None, [256], [0,256])
hist_r = cv.calcHist([im], [2], None, [256], [0,256])
plt.plot(np.log(hist_b), 'b')
plt.plot(np.log(hist_g), 'g')
plt.plot(np.log(hist_r), 'r')
plt.xlim([0,256])
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
im = cv.imread('C:/Users/Madara/Pictures/1.jpg', cv.IMREAD_GRAYSCALE)
hist = cv.calcHist([im], [0], None, [256], [0,256])
plt.plot(hist)

equ = cv.equalizeHist(im)

plt.xlim([0,256])
plt.show()

fig, ax = plt.subplots(1,2, figsize=(8,16))
ax[0].imshow(im, cmap='gray', vmin=0, vmax=255)
ax[0].xaxis.tick_top()
ax[1].imshow(equ, cmap='gray', vmin=0, vmax=255)
ax[1].xaxis.tick_top()
plt.show()
```

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
im = cv.imread('C:/Users/Madara/Pictures/1.jpg', cv.IMREAD_COLOR)
im = cv.cvtColor(im, cv.COLOR_BGR2RGB)
gamma = 2
t1 = np.array([(i/255)**gamma*255 for i in range (0, 256)]).astype(np.uint8)
im2 = t1[im]

gamma = 0.2
t2 = np.array([(i/255)**gamma*255 for i in range (0, 256)]).astype(np.uint8)
im3 = t2[im]

fig, ax = plt.subplots(1,3, figsize=(20,30))
ax[0].imshow(im)
ax[0].xaxis.tick_top()
ax[1].imshow(im2)
ax[1].xaxis.tick_top()
ax[2].imshow(im3)
ax[2].xaxis.tick_top()
plt.show()
plt.plot(t1, label='$\gamma = 2$')
plt.plot(t2, label='$\gamma = 0.22$')
plt.legend()
```
