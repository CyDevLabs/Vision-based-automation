```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**4 - x**3 -12*x**2 + 15*x + 5

def g(x):
    return 4*x**3 - 3*x**2 - 24*x +15

x0 = -3.9
alpha = 0.02
x_history = [x0]   
fx_history = [f(x0)]
x = x0
for i in range(20):
    x = x - alpha*g(x)
    x_history.append(x)
    fx_history.append(f(x))

x_ = np.arange(-4, 4 +0.01, 0.01)
plt.plot(x_, f(x_))
plt.scatter(x_history, fx_history, c='r')
plt.title(r'$x^\ast = {}, f(x^\ast) = {} $', format(x,f(x)))

```
